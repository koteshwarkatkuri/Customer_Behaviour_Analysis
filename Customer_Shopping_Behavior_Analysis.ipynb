```python
# Importing libraries
import numpy as np
import pandas as pd

# Loading the dataset
df=pd.read_csv("customer_shopping_behavior.csv")
df.head()
```

```python
# information of whole data
df.info()
```

```python
# Statistical information of data
print(df.describe())
print(df.describe(include="all"))
```

```python
# checking Null Values
df.isnull().sum()
```

```python
# fill null values in review rating columns with different category median which optimize fillinf missing values 
df['Review Rating']=df.groupby("Category")["Review Rating"].transform(lambda x:x.fillna(x.median()))
```

```python
# Rechecking the null values
df.isnull().sum()
```

```python
# Snake Casing Means converting all column names to lowercase
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ',"_")
df=df.rename(columns={"purchase_amount_(usd)":"purchase_amount"})
```

```python
df.columns
```

```python
# creating a new column  age_group
labels=["Young Adult","Adult","Middle-aged","Senior"]
df['age_group']=pd.qcut(df['age'],q=4,labels=labels) # qcut equals age into 4 groups based on data disribution and assigns the label
```

```python
df[['age','age_group']].head(10)
```

```python
# create column purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually' : 365,
    'Every 3 Months': 90
}
df["purchase_frequency_days"]=df["frequency_of_purchases"].map(frequency_mapping)
```

```python
df[['frequency_of_purchases','purchase_frequency_days']]
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
# checking both columns have same values  
(df['discount_applied']==df['promo_code_used']).all() 
```

```python
# Droping 1 column
df = df.drop('promo_code_used',axis=1)
```
